```python
def gaussian(x, m, s):
    return 1./(np.sqrt(2.*np.pi)*s)*np.exp(-np.power((x - m)/s, 2.)/2)
```

```python
x_values = np.linspace(-10, 10, 100) 
lines = []
for m, s in [(0,math.sqrt(1.5) ), (3, math.sqrt(4))]:
    plt.plot(x_values, gaussian(x_values, m, s))
    lines += plt.plot(x_values,  gaussian(x_values, m, s))

plt.legend(lines[:2], ['$\mathcal{N}(0,1.5)$', '$\mathcal{N}(3,4)$'],
          loc='upper right')

plt.xlabel('$x$', fontsize=13)
plt.ylabel('density: $p(x)$', fontsize=13)
plt.title('Univariate normal distributions')   
plt.grid()
plt.savefig('1a.png')
plt.show()
```
